```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('../DATA/digits.csv')
```

```python
digits
```

```python
pixels = digits.drop('number_label', axis=1)
```

```python
pixels
```

```python
pixels.iloc[0]
```

```python
arr = np.array(pixels.iloc[0])
```

```python
arr = arr.reshape(8,8)
arr
```

```python
plt.figure(figsize=(3,2), dpi=150)
sns.heatmap(arr, cmap='viridis', cbar=False)
```

```python
plt.figure(figsize=(3,2), dpi=150)
sns.heatmap(arr, cmap='viridis', annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(scaled_pixels)
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,6), dpi=150)
labels = digits['number_label']
sns.scatterplot(x=principal_components[:,0], y=principal_components[:,1], hue=labels, palette='Set1')
plt.legend(loc=(1.05,0))
```
